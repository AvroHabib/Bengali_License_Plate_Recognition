```python
import os
import shutil
import random
```

Transforming the dataset into YOLOV8 traininng format . This dataset is only for the OCR purpose.

```python
dataset_path = "/home/avrohabib/Downloads/Bangla LPDB - A/Bangla_License_Plate"  # Folder containing images and labels
output_path = "Only_License_Plate"  # Folder to save the images with only license plates
image_exts = [".jpg", ".png", ".jpeg"]
```

```python
os.makedirs(f"{output_path}/images/train", exist_ok=True)
os.makedirs(f"{output_path}/images/val", exist_ok=True)
os.makedirs(f"{output_path}/labels/train", exist_ok=True)
os.makedirs(f"{output_path}/labels/val", exist_ok=True)
```

```python
image_files = [f for f in os.listdir(dataset_path) if f.endswith(tuple(image_exts))]
random.shuffle(image_files)
```

Splitting dataset into 80%:20% for training and validation

```python
split_idx = int(0.8 * len(image_files))
```

```python
train_files = image_files[:split_idx]
val_files = image_files[split_idx:]
```

```python
for img_file in train_files:
    img_path = os.path.join(dataset_path, img_file)
    print(img_path)
    label_path = os.path.splitext(img_path)[0] + ".txt"
    print(label_path)

    if os.path.exists(label_path):
        shutil.move(img_path, f"{output_path}/images/train/")
        shutil.move(label_path, f"{output_path}/labels/train/")
```

```python
for img_file in val_files:
    img_path = os.path.join(dataset_path, img_file)
    label_path = os.path.splitext(img_path)[0] + ".txt"

    if os.path.exists(label_path):
        shutil.move(img_path, f"{output_path}/images/val/")
        shutil.move(label_path, f"{output_path}/labels/val/")
```
