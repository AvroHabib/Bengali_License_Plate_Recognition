```python
from PIL import Image

# Load the image
image_path = 'o4.jpg'
image = Image.open(image_path)

# Print out image properties
print(f"Format: {image.format}")
print(f"Size: {image.size}")
print(f"Mode: {image.mode}")
print(f"Info: {image.info}")
```

```python
import cv2
from matplotlib import pyplot as plt

# Load the image using OpenCV
image_cv = cv2.imread('o4.jpg')

# Print out image properties
print(f"Shape: {image_cv.shape}")
print(f"Size: {image_cv.size}")
print(f"Data type: {image_cv.dtype}")

# Resize the image to 640x640
resized_image = cv2.resize(image_cv, (100, 100))

# Convert the image from BGR to RGB (OpenCV loads images in BGR format)
resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)

# Display the resized image
plt.imshow(resized_image_rgb)
plt.axis('off')
plt.show()
```
